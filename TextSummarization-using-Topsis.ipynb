```python
import time
import pandas as pd
from transformers import pipeline
import evaluate  # Updated import for the ROUGE metric

# Define summarization models (now with 5 models)
models = [
    "facebook/bart-large-cnn",
    "t5-small",
    "google/pegasus-xsum",
    "facebook/bart-large-xsum",
    "google/pegasus-large"
]

# Sample text for summarization
text = """
Artificial Intelligence (AI) is transforming various industries by automating tasks, enhancing decision-making,
and improving efficiency. AI-powered systems can analyze vast amounts of data, recognize patterns, and make predictions,
which is beneficial in fields like healthcare, finance, and transportation. However, AI also raises ethical concerns
regarding data privacy, bias, and job displacement. Ensuring responsible AI development and usage is crucial for maximizing
its benefits while minimizing potential risks.
"""

# Load ROUGE metric using the new 'evaluate' library
rouge = evaluate.load("rouge")

# Dictionary to store model performance metrics
model_data = {
    "Model": [],
    "ROUGE-1": [],
    "ROUGE-2": [],
    "ROUGE-L": [],
    "Inference Time (ms)": [],
    "Model Size (MB)": [500, 300, 600, 500, 1000]  # Approximate model sizes for 5 models
}

for model_name in models:
    summarizer = pipeline("summarization", model=model_name)

    # Measure inference time
    start_time = time.time()
    summary = summarizer(text, max_length=50, min_length=20, do_sample=False)[0]["summary_text"]
    end_time = time.time()

    inference_time = (end_time - start_time) * 1000  # Convert to milliseconds

    # Compute ROUGE scores
    rouge_scores = rouge.compute(predictions=[summary], references=[text])

    # Store results
    model_data["Model"].append(model_name)
    model_data["ROUGE-1"].append(rouge_scores["rouge1"])  # Access fmeasure for each ROUGE score
    model_data["ROUGE-2"].append(rouge_scores["rouge2"])
    model_data["ROUGE-L"].append(rouge_scores["rougeL"])
    model_data["Inference Time (ms)"].append(round(inference_time, 2))

# Convert results to DataFrame
df_metrics = pd.DataFrame(model_data)
df_metrics
```

```python
# Define Weights (Adjust based on importance)
weights = [0.3, 0.3, 0.3, 0.1]  # Example: ROUGE-1, ROUGE-2, ROUGE-L, Inference Time

# Define Impacts (Maximize or Minimize)
impacts = ["+", "+", "+", "-"]  # ROUGE scores should be maximized, Inference Time minimized
```

```python
import numpy as np

def topsis(matrix, weights, impacts):
    """
    Perform TOPSIS ranking on a decision matrix.

    Parameters:
    - matrix: numpy array (MxN) where M is the number of alternatives and N is the number of criteria
    - weights: list of N weights
    - impacts: list of "+" or "-" for each criterion

    Returns:
    - topsis_scores: List of TOPSIS scores for each alternative
    - rankings: List of ranks (1 = best)
    """

    # Convert to numpy array
    matrix = np.array(matrix, dtype=float)
    weights = np.array(weights, dtype=float)

    # Step 1: Normalize the decision matrix
    norm_matrix = matrix / np.sqrt((matrix ** 2).sum(axis=0))

    # Step 2: Multiply by weights
    weighted_matrix = norm_matrix * weights

    # Step 3: Identify ideal best and worst values
    ideal_best = np.zeros(weighted_matrix.shape[1])
    ideal_worst = np.zeros(weighted_matrix.shape[1])

    for i in range(weighted_matrix.shape[1]):
        if impacts[i] == "+":
            ideal_best[i] = np.max(weighted_matrix[:, i])  # Max for benefit criteria
            ideal_worst[i] = np.min(weighted_matrix[:, i])  # Min for benefit criteria
        else:
            ideal_best[i] = np.min(weighted_matrix[:, i])  # Min for cost criteria
            ideal_worst[i] = np.max(weighted_matrix[:, i])  # Max for cost criteria

    # Step 4: Calculate distances to ideal best and worst
    dist_best = np.sqrt(((weighted_matrix - ideal_best) ** 2).sum(axis=1))
    dist_worst = np.sqrt(((weighted_matrix - ideal_worst) ** 2).sum(axis=1))

    # Step 5: Compute TOPSIS score
    topsis_scores = dist_worst / (dist_best + dist_worst)

    # Step 6: Rank based on TOPSIS score (higher is better)
    rankings = np.argsort(-topsis_scores) + 1  # Rank 1 = Best

    return topsis_scores, rankings
```

```python
import pandas as pd

# Define metrics columns
metrics_columns = ['ROUGE-1', 'ROUGE-2', 'ROUGE-L', 'Inference Time (ms)']

# Define weights & impacts
weights = [0.3, 0.3, 0.3, 0.1]  # Adjust as needed
impacts = ["+", "+", "+", "-"]

# Ensure DataFrame contains required columns
assert all(col in df_metrics.columns for col in metrics_columns), "Some required columns are missing!"

# Apply TOPSIS
df_metrics['TOPSIS Score'], df_metrics['Rank'] = topsis(df_metrics[metrics_columns].values, weights, impacts)

# Sort by TOPSIS Score and Rank
df_metrics = df_metrics.sort_values(by=['TOPSIS Score', 'Rank'], ascending=[False, True], ignore_index=True)

# Save results
df_metrics.to_csv("topsis_summarization_ranking.csv", index=False)

# Print results
print("\n✅ Ranked Summarization Models Based on TOPSIS:")
print(df_metrics[['TOPSIS Score', 'Rank']].head(10))
```
